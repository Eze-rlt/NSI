Correction exercice 14:

```python
def examen(x,y,z,t):
    '''cette fonction permet de calculer la moyenne de 4 notes dont les 2 premieres sont coefficient 2'''
    assert type(x)==float or type(x)==int, "ne rentrer que des nombre"
    assert type(y)==float or type(y)==int, "ne rentrer que des nombre"
    assert type(z)==float or type(z)==int, "ne rentrer que des nombre"
    assert type(t)==float or type(t)==int, "ne rentrer que des nombre"
    m=(2*x+2*y+z+t)/6
    if m>=10:
        print("le candidat est reçu")
    else:
        print("le condidat est refusé")
    return None
```

```python
examen(12,14.5,7,6)
```

Correction exercice 16:

```python
def get_max (liste):
    max = liste[0]
    for i in range (1,len(liste)):
        if liste[i] > max:
            max = liste[i]
    return (max)
```
